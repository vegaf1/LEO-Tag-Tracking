```julia
using LinearAlgebra 
using ForwardDiff
using Plots
using BlockDiagonals

#Speed of light
#c = 299792.458 # in km/s

time_scale = 1/45.679
distance_scale = 6563

c = 1 #distance_unit/time_unit

#Satellite Positions in kilometersmeters
#contain x,y,z,t (receive_time)
#r1 = [0,0,7370, 0.0033356409519815205]
#r2 = [7370,2000,8370, 0.02633190123748905]
#r3 = [-7370,7370,9370, 0.03617806713518567]
#r4 = [4000,-4000,10370, 0.023109998418557643]

#r1 = [-3000,-3000, 7000, 0.01530320506503748]
#r2 = [7370,2000,7370, 0.026686810023557543]
#r3 = [-7370,7370,8370, 0.0363958288914586]
#r4 = [4000,-4000,7970, 0.020602237960340566]

#best test (around 10's)
r1 = [-100,-100, 11500, 0.01809166409519016]
r2 = [7370,2000,8370, 0.02732515312128262]
r3 = [-7370,7370,9370, 0.037170695068067625]
r4 = [4000,-4000,10370, 0.02409460203090537]

#Testing orbit positions
#r1 = [6610.963224782449, 26.944111955229776, 1285.3957697661945, 0.005583450220303326]
#r2 = [5692.4408074083885, 2762.5160618596638, -2310.7752880133503, 0.012985102280545592]
#r3 = [2720.855721136459, 4440.200781545761, -4280.4520212476245, 0.024373634626189123]
#r4 = [-24.85731557858654, 6616.7203168928, -1289.1479118132336, 0.031265348297013384]


#best test (around 10's)
#r1 = [0,0, 11500, 0.01808515295604935]
#r2 = [7370,2000,8370, 0.02732515312128262]
#r3 = [-7370,7370,9370, 0.037170695068067625]
#r4 = [4000,-4000,10370, 0.02409460203090537]

#Lowest error
#r1 = [0,0, 11500, 0.0181118380836652]
#r2 = [7370,2000,8370, 0.02733190123748905]
#r3 = [-7370,7370,9370, 0.03717806713518567]
#r4 = [4000,-4000,10370, 0.024109998418557644]

# 1 km case
#r1 = [1,4,7370, 0.004335669304809114]
#r2 = [0,3,7370.8, 0.00433832446309499]
#r3 = [-0.75,0,7368, 0.004328970610106501]
#r4 = [-1,4,7371, 0.004339004917436832]

#100 km case (LEO)
#r1 = [-100,0,7385, 0.004402067613317489]
#r2 = [100,0,7375, 0.004368873511169709]
#r3 = [0,100,7380,  0.0043854701634201895]
#r4 = [0, -100,7383, 0.004395428538389553]

#10000 km case (MEO) 
#r1 = [-10000,0,7385, 0.03452779227884826]
#r2 = [10000,0,7375, 0.03452444033510887]
#r3 = [0,10000,7380, 0.034526112200424036]
#r4 = [0, -10000,7385, 0.03452779227884826]

#100 km sphere
#r1 = [6470,10,10, 0.031262665729212775]
#r2 = [-6470,20,20,  0.031239346831024535]
#r3 = [5, 6470, 5, 0.031274346006952676]
#r4 = [2,-6470, 2,  0.03128136082951181]

#different pose
#r1 = [6470,10,10, 0.03130946996205797]
#r2 = [-6470,20,20,  0.031332955039900794]
#r3 = [5, 6470, 5, 0.0312977481394262]
#r4 = [2,-6470, 2, 0.031290721684298924]

all_sats = vcat(r1',r2',r3',r4')
all_sats_scaled = [all_sats[:,1:3]/distance_scale all_sats[:,4]/time_scale]

print(all_sats)
```

```julia
function residual(x, sat_pose_t)
    
    return [norm(sat_pose_t[1,1:3] - x[1:3]) - c*(sat_pose_t[1,4] - x[4]);
            norm(sat_pose_t[2,1:3] - x[1:3]) - c*(sat_pose_t[2,4] - x[4]);
            norm(sat_pose_t[3,1:3] - x[1:3]) - c*(sat_pose_t[3,4] - x[4]);
            norm(sat_pose_t[4,1:3] - x[1:3]) - c*(sat_pose_t[4,4] - x[4])]
end
```

```julia
n = 1000 # number of iterations
all_r0 = NaN*[zeros(4) for i = 1:n]

sat1poses = zeros(4,n)
sat2poses = zeros(4,n)
sat3poses = zeros(4,n)
sat4poses = zeros(4,n)

all_sats_noise = zeros(4,4)


iters = 0

for i in 1:n
    #create noise from random distribution
    #scaled to the variable (1m error for distance & 1e-9 for time)
    gpsnoise = randn(12)*0.1*(1e-3/distance_scale) #adding a meter of noise
    clocknoise = randn(4)*(1e-9/time_scale)
    iterations = 0
    
    sat1_noise = [all_sats_scaled[1,1]+gpsnoise[1],all_sats_scaled[1,2] + gpsnoise[2],all_sats_scaled[1,3]+gpsnoise[3], all_sats_scaled[1,4] + clocknoise[1]]
    sat2_noise =[all_sats_scaled[2,1]+gpsnoise[4],all_sats_scaled[2,2] + gpsnoise[5],all_sats_scaled[2,3]+gpsnoise[6], all_sats_scaled[2,4] + clocknoise[2]]
    sat3_noise = [all_sats_scaled[3,1]+gpsnoise[7],all_sats_scaled[3,2] + gpsnoise[8],all_sats_scaled[3,3]+gpsnoise[9], all_sats_scaled[3,4] + clocknoise[3]]
    sat4_noise = [all_sats_scaled[4,1]+gpsnoise[10],all_sats_scaled[4,2] + gpsnoise[11],all_sats_scaled[4,3]+gpsnoise[12], all_sats_scaled[4,4] + clocknoise[4]]
                     
    all_sats_noise[:,:] = vcat(sat1_noise',sat2_noise',sat3_noise',sat4_noise')
    
    sat1poses[:,i] = sat1_noise
    sat2poses[:,i] = sat2_noise
    sat3poses[:,i] = sat3_noise
    sat4poses[:,i] = sat4_noise
    
    X = NaN*[zeros(4) for i = 1:10]
    R = NaN*[zeros(4) for i = 1:10]

    #Initial guess r0
    #X[1] = [100/distance_scale, 100/distance_scale, 6570/distance_scale, 0.1/time_scale]
    #usual guess for 0,0,6370
    X[1] = [1000/distance_scale, 1000/distance_scale, 5000/distance_scale, 0.1/time_scale]
    #guess for actual orbit positions
    #X[1] = [10/distance_scale, 10/distance_scale, 6300/distance_scale, 0.1/time_scale]

    #Initial residual
    R[1] = residual(X[1], all_sats_noise)

    iters = 0
    
    #Monte Carlo Simulation
    for k=1:20
        
        R[k] = residual(X[k], all_sats_noise)

        iters += 1

        if(norm(R[k]) < 1e-10)

            break

        end

        jacobian = ForwardDiff.jacobian(dx -> residual(dx, all_sats_noise), X[k])
        
        #print("this is the jacobian: ", jacobian)
        
        #print("these are the eigen values: ", eigvals(jacobian))
        
        conditionnum = cond(jacobian)

        deltax = jacobian\-R[k]

        X[k+1] = X[k] + deltax

    end  

    R = R[1:iters]
    X = X[1:iters]
    all_r0[i] = X[end]
    
end
```

```julia
#Least Squares Solution

function LS_Amatrix(x0, r_sat)
    jacobian_pose = (x0[1:3] - r_sat[1:3])/norm(r_sat[1:3]-x0[1:3])
    dpdtau = c
    
    return [jacobian_pose; dpdtau]
end

#Initial Guess
#x0 = [1000/distance_scale ,2200/distance_scale, 4360/distance_scale, 0.1/time_scale]
x0 = [0/distance_scale , 0/distance_scale, 6300/distance_scale, 0.1/time_scale]
```

```julia
function custom_jacobian(all_sats_scaled, x0)

    while norm(residual(x0, all_sats_scaled))>1e-10

        P_computed1 = norm(all_sats_scaled[1,1:3]-x0[1:3]) + c*x0[4] - c*all_sats_scaled[1,4]
        P_computed2 = norm(all_sats_scaled[2,1:3]-x0[1:3]) + c*x0[4] - c*all_sats_scaled[2,4]
        P_computed3 = norm(all_sats_scaled[3,1:3]-x0[1:3]) + c*x0[4] - c*all_sats_scaled[3,4]
        P_computed4 = norm(all_sats_scaled[4,1:3]-x0[1:3]) + c*x0[4] - c*all_sats_scaled[4,4]

        P_computed = [P_computed1, P_computed2, P_computed3, P_computed4]
        P_observed = zeros(4)
        deltaP = P_observed - P_computed

        A = vcat(LS_Amatrix(x0, all_sats_scaled[1,:])', LS_Amatrix(x0, all_sats_scaled[2,:])', LS_Amatrix(x0, all_sats_scaled[3,:])', LS_Amatrix(x0, all_sats_scaled[4,:])')

        F = qr(A)

        d = transpose(F.Q)*deltaP

        deltax = F.R\d

        x0 = x0+deltax
    end

    x_sol = x0
    x_sol_rescaled = [x_sol[1:3]*distance_scale, x_sol[4]*time_scale]
end
```

```julia
function Forwarddiffjacobian(all_sats_scaled, x0)
    
    X = NaN*[zeros(4) for i = 1:100]
    R = NaN*[zeros(4) for i = 1:100]

    X[1] = x0
    iters = 0
    
    
    for k=1:100

        R[k] = residual(X[k], all_sats_scaled)

        iters += 1

        if(norm(R[k]) < 1e-10)

            break

        end

        jacobian = ForwardDiff.jacobian(dx -> residual(dx, all_sats_scaled), X[k])

        #print("this is the jacobian: ", jacobian)

        #print("these are the eigen values: ", eigvals(jacobian))

        conditionnum = cond(jacobian)

        deltax = jacobian\-R[k]

        X[k+1] = X[k] + deltax

    end

    R = R[1:iters]
    X = X[1:iters]
    x_solution = X[end]
    x_solution_scaled = [x_solution[1]*distance_scale, x_solution[2]*distance_scale, x_solution[3]*distance_scale, x_solution[4]*time_scale]

end
    
# #while norm(residual(x0, all_sats_scaled))>1e-10
    
#     P_computed1 = norm(all_sats_scaled[1,1:3]-x0[1:3]) + c*x0[4] - c*all_sats_scaled[1,4]
#     P_computed2 = norm(all_sats_scaled[2,1:3]-x0[1:3]) + c*x0[4] - c*all_sats_scaled[2,4]
#     P_computed3 = norm(all_sats_scaled[3,1:3]-x0[1:3]) + c*x0[4] - c*all_sats_scaled[3,4]
#     P_computed4 = norm(all_sats_scaled[4,1:3]-x0[1:3]) + c*x0[4] - c*all_sats_scaled[4,4]

#     P_computed = [P_computed1, P_computed2, P_computed3, P_computed4]
#     P_observed = zeros(4)
#     deltaP = P_observed - P_computed
    
#     A = vcat(LS_Amatrix(x0, all_sats_scaled[1,:])', LS_Amatrix(x0, all_sats_scaled[2,:])', LS_Amatrix(x0, all_sats_scaled[3,:])', LS_Amatrix(x0, all_sats_scaled[4,:])')
    
#     F = qr(A)
    
#     d = transpose(F.Q)*deltaP
    
#     deltax = F.R\d
    
#     x0 = x0+deltax
# end

# x_sol = x0
# x_sol_rescaled = [x_sol[1:3]*distance_scale, x_sol[4]*time_scale]
```
